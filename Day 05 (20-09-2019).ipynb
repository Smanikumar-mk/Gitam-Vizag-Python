# idiomatic python 
- transformation code into beautiful,idomatic python
- replaces traditional index manipulation with python code looping idioms

```python
#looping over a raange of number
for i in[0,1,2,3,4,5]:
    print(i ** 2,end=' ')
    
    
print()
#pythinic-way
for i in range(5):
    print(i ** 2,end=' ')
```

```python
#looping over the collection of list

l=[1,2,3,4,5]
for i in range(len(l)):
    print(l[i],end=' ')
print()
#pythinic-way
for i in l:
    print(i,end=' ')
    
```

```python
# looping  from the backside

l=[1,2,3,4,5]
for i in range(len(l)-1,-1,-1):
     print(l[i],end=' ')
print()
#pythinic-way
for i in reversed(l):
     print(i,end=' ')
```

```python
# looping over the collection with index
li = [1,2,3,4,5]
for i in range(len(li)):
    print(i,'---->',li[i])
    print()
#pythonic --way
for i,a in enumerate(li):
    print(i,'---->',a)
```

```python

# looping over two collections
li = [1,2,3,4,5]
a  = [1,2,3,4,5,6,8]
n  = min(len(li),len(a))
for i in range(n):
    print(li[i],' ',a[i])
print()
#pythonic --way
for i,a in zip(li,a):
    print(i,' ',a)
```

### Lambda with reduce()


```python
def sumlist(l):
    s = 0
    for i in l:
        s += i
    return s
l=[1,2,3,4,5,6,7,8]
sumlist(l)
```

```python
from functools import reduce
l=[1,2,3,4,5,6,7,8]
s = reduce((lambda x,y:x+y),l)
print(s)
```
