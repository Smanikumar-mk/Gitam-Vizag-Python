## visualizing the data 
  - an untersting way to diisplay datafraame
  - with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
marks = [36,37,38,39]
avg_marks = [46,48,49,50]
divisions = ['A','B','C','D']
index = np.arange(len(divisions))
plt.bar(divisions,marks,color='green',width =0.375)
plt.show()
```

```python
plt.barh(divisions,marks,color='pink')
plt.show()
```

```python
apps = ['Free apps','affordable apps','expensive apps','non-expensive apps']
rating =[ 0,20,50,50]
plt.pie(rating,labels = apps,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
